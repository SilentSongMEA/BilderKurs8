```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic2\004.JPG")
```

```python
point_a=np.float32([[0,60],[150,20],[80,0],[210,140]])
point_b=np.float32([[0,50],[140,10],[70,0],[200,110]])
m=cv2.getPerspectiveTransform(point_a,point_b)
warped=cv2.warpPerspective(image,m,(300,300))
cv2.imshow("result",warped)
cv2.waitKey()
cv2.destroyAllWindows()
plt.figure(figsize=[10,5])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("input");
plt.subplot(122);plt.imshow(warped[...,::-1]);plt.title("Output");
```
