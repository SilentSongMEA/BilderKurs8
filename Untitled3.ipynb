```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
gray_img=cv2.imread(r"C:\\Users\Student\Desktop\pic2\001.jpeg",0)
gray_blur=cv2.blur(gray_img,(3,3))
sobel_x=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobel_y=np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
edge_x=cv2.filter2D(gray_blur,cv2.CV_32F,sobel_x)
edge_y=cv2.filter2D(gray_blur,cv2.CV_32F,sobel_y)
plt.figure(figsize=[17,8])
plt.subplot(131);plt.imshow(gray_img,cmap="gray");plt.title("orginal");
plt.subplot(132);plt.imshow(edge_x,cmap="gray");plt.title("sobel_x");
plt.subplot(133);plt.imshow(edge_y,cmap="gray");plt.title("sobel_y");
```

```python
orian=cv2.phase(np.array(edge_x,np.float32),np.array(edge_y,dtype=np.float32),angleInDegrees=True)
plt.figure(figsize=[17,8])
plt.imshow(orian,cmap="tab10"),plt.colorbar(),plt.show();
```

```python
hog=cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
image=cv2.imread(r"C:\\Users\Student\Desktop\pic2\002.jpeg")
bounding_box,weights=hog.detectMultiScale(image,winStride=(4,4),scale=1.05)
for (x,y,w,h) in bounding_box:
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,255),4)
plt.imshow(image[...,::-1])
```
