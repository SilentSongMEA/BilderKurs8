```python
img1=cv2.imread(r"C:\\Users\Student\Desktop\pic2\box.png")
img2=cv2.imread(r"C:\\Users\Student\Desktop\pic2\box_in_scene.png")
plt.figure(figsize=[10,5])
plt.subplot(121);plt.imshow(img1[...,::-1]);plt.title("image1");
plt.subplot(122);plt.imshow(img2[...,::-1]);plt.title("image2");
```

```python
sift=cv2.SIFT_create()
kp1,des1=sift.detectAndCompute(img1,None)
kp2,des2=sift.detectAndCompute(img2,None)
print(des1.shape)
print(des2.shape)
```

```python
keypoints_without_size=np.copy(img1)
keypoints_with_size=np.copy(img1)
cv2.drawKeypoints(img1,kp1,keypoints_without_size,color=(0,255,0))
cv2.drawKeypoints(img1,kp1,keypoints_with_size,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.figure(figsize=[10,5])
plt.subplot(121),plt.imshow(keypoints_without_size[...,::-1])
plt.subplot(122),plt.imshow(keypoints_with_size[...,::-1])
```

```python
bf=cv2.BFMatcher(cv2.NORM_L1,crossCheck=False)
matches=bf.match(des1,des2)
matches=sorted(matches,key=lambda x:x.distance)
img3=cv2.drawMatches(img1,kp1,img2,kp2,matches[:20],None,flags=
                    cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
plt.imshow(img3),plt.show()
print("number of matching keypoints",len(matches))
```

```python
img=cv2.imread(r"C:\\Users\Student\Desktop\pic2\006.jpeg")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray=np.float32(gray)
dst=cv2.cornerHarris(gray,2,3,0.04)
dst=cv2.dilate(dst,None,iterations=1)
img[dst>0.01*dst.max()]=[255,0,255]
plt.imshow(img,cmap="gray")
```

```python
img=cv2.imread(r"C:\\Users\Student\Desktop\pic2\006.jpeg")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
corners=cv2.goodFeaturesToTrack(gray,7,0.07,10)
corners=np.int0(corners)
for i in corners:
    x,y=i.ravel()
    cv2.circle(img,(x,y),3,255,-1)
plt.imshow(img),plt.show();
```
